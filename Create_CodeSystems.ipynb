```python
import pandas as pd
```

# England from Website

```python
#Choose site with table
data = pd.read_html('https://www.datadictionary.nhs.uk/attributes/main_specialty_code.html')
```

```python
#Choose table within list
df = data[0]
df
```

```python
# Create etxt file For England
text_file = open("MainSpecialityEngland.txt","w")
for row in df.iterrows():
    if 'retired' in row[1][1]:
        pass
    else:
        text_file.write('<concept>\n\t<code value = "'+str(row[1][0])+'" />\n\t<display value = "'+row[1][1]+'" />\n</concept>\n')
        
```

# Wales from csv


```python
csvdata = pd.read_csv('scotland.csv',header=None)
```

```python
csvdata
```

```python
# remove any NaN from 2nd column
csvdata.dropna(subset=[csvdata.columns[1]],inplace=True)
```

```python
#remove any retired entries (columns 4)
csv_data = csvdata[csvdata[csvdata.columns[3]].isnull()]
csv_data
```

```python
#Create txt file For Scotland
text_file = open("MainSpecialityWales.txt","w")
for row in csv_data.iterrows():
    text_file.write('<concept>\n\t<code value = "'+str(row[1][0])+'" />\n\t<display value = "'+row[1][1]+'" />\n</concept>\n')
        
```
